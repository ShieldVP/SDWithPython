# Matplotlib

```python
import matplotlib.pyplot as plt
```

Библиотека **Matplotlib** - это библиотека визуализации данных для Python, которая предоставляет множество инструментов для создания разнообразных графиков и диаграмм.<br>
Основные фичи matplotlib:

1. **Создание разнообразных графиков**: Matplotlib позволяет создавать графику различных типов, включая линейные графики, точечные графики, гистограммы, круговые диаграммы, столбчатые диаграммы и многое другое.

2. **Настройка графиков**: Вы можете настраивать почти все аспекты графиков, включая заголовки, подписи осей, цвета, стили линий, маркеры и многое другое.

3. **Интерактивность**: Matplotlib поддерживает интерактивное взаимодействие с графиками, что позволяет пользователю взаимодействовать с данными и элементами графика.

4. **Сохранение в различных форматах**: Вы можете сохранять графики в разных форматах, таких как PNG, JPEG, PDF, SVG и других.

5. **Визуализация данных**: Matplotlib используется для визуализации данных, что делает его незаменимым инструментом для анализа данных и презентации результатов.

6. **Интеграция с другими библиотеками**: Matplotlib хорошо интегрируется с другими библиотеками для анализа данных в Python, такими как NumPy и pandas.

7. **Поддержка нескольких интерфейсов**: Вы можете использовать Matplotlib в разных интерфейсах, включая стандартный интерфейс для создания графиков из командной строки и интерактивные интерфейсы для создания графиков в оконном режиме.

8. **Гибкая настройка структуры графика**: Вы можете создавать мультиграфики (несколько графиков на одной области), вложенные графики и настраивать структуру графика по своему усмотрению.

9. **Создание анимаций**: Matplotlib позволяет создавать анимации, что полезно для демонстрации изменения данных во времени.

10. **Темы и стили**: Вы можете легко настраивать внешний вид графиков с помощью предопределенных тем и стилей, или создавать свои собственные.

Matplotlib является одной из наиболее популярных библиотек для визуализации данных в Python и широко используется в научных и инженерных приложениях, а также для создания графиков для отчетов и презентаций.

# Создание графиков

```python
# настройка размеров графика в интерактивном режиме
plt.rcParams ['figure.figsize'] = [15, 10]
```

```python
import numpy as np

# Создание линейного графика
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('График синуса')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')

# при работе в jupyter не обязательно писать plt.show()
plt.show()
```

# Настройка графиков

```python
# Настраиваем график: цвет, стиль линии и маркеры
plt.plot(x, y, color='blue', linestyle='--', marker='.', label='Синус')
plt.title('График синуса')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.legend()
plt.grid(True)
plt.show()
```

# Визуализация данных

```python
import pandas as pd

# Визуализация данных из DataFrame
data = {'Год': [2010, 2015, 2020],
        'Продажи': [100, 150, 200]}
df = pd.DataFrame(data)
plt.bar(df['Год'], df['Продажи'])
plt.title('Динамика продаж')
plt.xlabel('Год')
plt.ylabel('Продажи')
plt.show()
```

# Настройка визуализации

```python
# Создание мультиграфика
plt.figure(figsize=(10, 4))

# Первый график
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x), label='Синус')
plt.title('График синуса')

# Второй график
plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x), label='Косинус', color='red')
plt.title('График косинуса')

plt.tight_layout()  # Улучшает расположение графиков на странице
plt.show()
```

# Создание анимаций

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()

def animate(i):
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(x + i * 0.1)
    ax.clear()
    ax.plot(x, y)
    ax.set_title(f'Анимация фазы: {i * 0.1:.2f}')

ani = animation.FuncAnimation(fig, animate, frames=30, interval=200)
ani.save('animation.gif', writer='pillow')
```
